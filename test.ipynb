```python
# load libraries & init notebook
import re
import pandas as pd
import numpy as np
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
```

```python
# build initial dataframe
df = pd.read_csv('notas-vd-2024-25.csv', sep='\t', header=0)
year = "2024-25"
```

```python
# initial scatterplot
data = []
data.append(go.Scatter(x=df['ec'], y=df['examen'], text=df['id'], mode='markers'))
fig = go.Figure(layout_title_text=f"Notas VD {year}: evaluación contínua vs examen final", data=data)
fig.update_xaxes(range=[0, 10])
fig.update_yaxes(range=[0, 10])
fig.update_layout(
    xaxis_title="nota evaluación continua (40% total)",
    yaxis_title="nota examen final (60% total) ",
    width=600, height=600)
fig.show()
```

```python
# build histograms
data = []
data.append(go.Histogram(x=df['ec'], text=df['id']))
fig = go.Figure(layout_title_text=f"Notas VD {year}: evaluación contínua", data=data)
fig.update_xaxes(range=[-1, 11])
fig.update_yaxes(range=[0, 10])
fig.update_layout(
    xaxis_title="nota evaluación continua (40% total)",
    yaxis_title="alumnos",
    width=600, height=300)
fig.show()

data = []
data.append(go.Histogram(x=df['examen'], text=df['id']))
fig = go.Figure(layout_title_text="Notas VD {year}: examen final", data=data)
fig.update_xaxes(range=[-1, 11])
fig.update_yaxes(range=[0, 10])
fig.update_layout(
    xaxis_title="nota examen final (60% total)",
    yaxis_title="alumnos",
    width=600, height=300)
fig.show()
```

```python
# todo junto (con subplots)

fig = make_subplots(rows=2, cols=2, subplot_titles=['', 'nota evaluación contínua (40% total)', '', ''],
    row_heights=[0.3, 0.7], column_widths=[0.3, 0.7],
    horizontal_spacing=0.05, vertical_spacing=0.05)
fig.update_yaxes(range=[0, 10], tick0=0, dtick=1)
fig.update_xaxes(range=[0, 10], tick0=0, dtick=1)

scatter = go.Scatter(x=df['ec'], y=df['examen'], text=df['id'], mode='markers', name="ec vs. examen")
fig.add_trace(scatter, row=2, col=2)

ec_histogram = go.Histogram(x=df['ec'], name="EC", xbins=dict(size=1, start=0, end=10))
fig.add_trace(ec_histogram, row=1, col=2)
fig.update_yaxes(range=[0, None], row=1, col=2, title_text="# alumnos")
                 
ex_histogram = go.Histogram(y=df['examen'], name='examen final', ybins=dict(size=1, start=0, end=10))
fig.update_yaxes(title_text="nota examen final (60% total)", row=2, col=1)
fig.add_trace(ex_histogram, row=2, col=1)
fig.update_xaxes(autorange='reversed', row=2, col=1, title_text="# alumnos")

fig.update_layout(width=1000, height=800, showlegend=False, title_text=f"Notas VD {year}: evaluación contínua vs examen final")
fig.show()
```

```python
# final grades
data = []
data.append(go.Histogram(x=df['final'], text=df['id'], xbins=dict(size=1, start=0, end=10), ))
fig = go.Figure(layout_title_text=f"Notas VD {year}: nota final (mayo)", data=data)
fig.update_xaxes(title_text="nota final", dtick=1)
fig.update_layout(
    yaxis_title="alumnos",
    width=800, height=400)
fig.show()
```
